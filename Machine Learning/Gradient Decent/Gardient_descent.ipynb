### Gradient Descent:

#### Gradient descent is an algogrithm that finds best fit line for given training data set.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
def grad_dec(x, y):
    
    m = b = 0
    iterations =100
    learning_rate = 0.007
    n = len(x)
    plt.scatter(x,y,color='red', marker='*',linewidth=5)
    for i in range(iterations):
        yp = m * x + b
        plt.plot(x,yp, color='green')
        cost = (1/n)*sum([val**2 for val in (y-yp)])
        pdm = -(2/n) * sum(x*(y-yp))
        pdb = -(2/n) * sum(y-yp)
        m = m - learning_rate * pdm
        b = m - learning_rate * pdb
        print('m {}, b {}, cost {}, iteration {}'.format(m, b, cost, i))
    
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
grad_dec(x, y)
```
